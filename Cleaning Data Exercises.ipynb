## Missing Data

In the Towards Data Science article, they list 3 techniques for detecting any missing data frame. Try all three before deciding on a way to solve any missing data issues.

```python
# Plot a heatmap of any missing data
cols = NYRat.columns
colours = ['#000099', '#ffff00'] # specify the colours - yellow is missing. blue is not missing.
sns.heatmap(NYRat[cols].isnull(), cmap=sns.color_palette(colours))
```

```python
# Try out a missing data percentage list! 
for col in NYRat.columns:
    pct_missing = np.mean(NYRat[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Plot a missing data histogram
for col in NYRat.columns:
    missing = NYRat[col].isnull()
    num_missing = np.sum(missing)
    
    if num_missing > 0:  
        print('created missing indicator for: {}'.format(col))
        NYRat['{}_ismissing'.format(col)] = missing

ismissing_cols = [col for col in NYRat.columns if 'ismissing' in col]
NYRat['num_missing'] = NYRat[ismissing_cols].sum(axis=1)

NYRat['num_missing'].value_counts().reset_index().sort_values(by='index').plot.bar(x='index', y='num_missing')
```

## Unnecessary Data

Unnecessary data could be duplicates, irrelevant, or any uninformative data. Dive into each type of unnecessary data to see if there is any in the data set. Make sure to make note of how you would handle each type of unnecessary data.

```python
# Look for any irrelevant data in the dataframe. How will you handle it?
cols_not_needed=[]
for col in NYRat_Clean5.columns:
    if '_ismissing' in col:
        cols_not_needed.append(col)
    else:
        continue
    
print(cols_not_needed)
print(len(cols_not_needed))
```

```python
# Think about what data in the dataframe may be uninformative. Make note here of what situations may render data uninformative and how you would handle it?
cols_not_unique=[]
for col in NYRat_Clean7.columns:
    if len(pd.unique(NYRat_Clean7[col]))<=1:
        cols_not_unique.append(col)
    else:
        continue
        
print(cols_not_unique)
print(len(cols_not_unique))
```

## Inconsistent Data

Inconsistent data is anything that messes with your model. This is likely due to inconsistent formatting and can be addressed by re-formatting all values in a column or row.

```python
# Try to convert all the strings in a column to lower or upper case. 
NYRat_Clean8['Incident Address'] = NYRat_Clean8['Incident Address'].str.lower()
NYRat_Clean8['Incident Address'] = NYRat_Clean8['Incident Address'].str.strip() # remove leading and trailing whitespace.
NYRat_Clean8['Incident Address'] = NYRat_Clean8['Incident Address'].str.replace('\\.', '', regex=True) # remove period.
NYRat_Clean8['Incident Address'] = NYRat_Clean8['Incident Address'].str.replace('\\bstreet\\b', 'st', regex=True) # replace street with st.
NYRat_Clean8['Incident Address'] = NYRat_Clean8['Incident Address'].str.replace('\\bboulevard\\b', 'blvd', regex=True) # replace street with st.
NYRat_Clean8['Incident Address'] = NYRat_Clean8['Incident Address'].str.replace('\\bapartment\\b', 'apt', regex=True) # replace apartment with apt.
NYRat_Clean8['Incident Address'] = NYRat_Clean8['Incident Address'].str.replace('\\bav\\b', 'ave', regex=True) # replace av with ave.
NYRat_Clean8['Incident Address'] = NYRat_Clean8['Incident Address'].str.replace('\\bavenue\\b', 'ave', regex=True) # replace avenue with ave.

NYRat_Clean8['Incident Address'].head()
```
